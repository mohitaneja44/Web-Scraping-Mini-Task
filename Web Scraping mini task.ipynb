```python
import bs4  #Importing the BeautifulSoup library

from urllib import request
```

```python
from urllib.request import urlopen 

from bs4 import BeautifulSoup 
```

```python
list_of_urls = ['https://www.seedrs.com/','https://www.crowdcube.com/', 'https://www.fundingcircle.com/us/']
```

```python
list_of_urls[0]
```

```python
things = []
faulty_URLs = []
for urls in list_of_urls:
    try:
        things.append(urlopen(urls))
    except:
        print("Some error while handling the URL: {}".format(urls))
        faulty_URLs.append(urls.split(".")[1])
```

```python
things
```

```python
list_of_page_html = []
for thing in things:

    list_of_page_html.append(thing.read())  #Reading the page
    thing.close()
print(list_of_page_html)
```

```python
list_of_soups = []
text_list = []

for veggies in list_of_page_html:
    soup_of_data = BeautifulSoup((veggies) , "html.parser") #Parsing the HTML
    text_list.append(soup_of_data.get_text()) #appending the text data into a list
```

```python
names_of_URLs = [ url.split(".")[1] for url in list_of_urls]
names_of_URLs
```

```python
Good_URLs = [x for x in names_of_URLs if x not in faulty_URLs] #Removing the faulty URLs from the list of URLs 
Good_URLs
```

```python
import io 

for i in range(len(Good_URLs)):  #Writing the data into a text file
    with io.open( Good_URLs[i] + '.txt', 'w', encoding='utf8') as file: #decoding the unicode encoding of the text
        file.write(text_list[i])
```
